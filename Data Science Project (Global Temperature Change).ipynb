```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("allcountriesglobaltemperaturechange.csv")
```

```python
# Filling the BLank data using interpolation method 
year_columns = df.columns[4:]  
df[year_columns] = df[year_columns].apply(pd.to_numeric, errors='coerce')
df[year_columns] = df[year_columns].interpolate(method='linear', axis=1, limit_direction='both')
```

```python
df.isnull().sum()
```

```python
year_cols = [str(year) for year in range(1970, 2022)]
```

```python
# Country with Highest rise of Temperature from baseline
max_val = df[year_cols].max().max()
max_year = df[year_cols].max().idxmax()   
max_country = df.loc[df[max_year].idxmax(), "Country Name"]
print("Highest rise value:", max_val)
print("Year:", max_year)
print("Country:", max_country)
```

```python
# Counrty with lowest drop of temperature form baseline
min_val = df[year_cols].min().min()
min_year = df[year_cols].min().idxmin()
min_country = df.loc[df[min_year].idxmin(), "Country Name"]
print("lowest drop value:", min_val)
print("Year:", min_year)
print("Country:", min_country)
```

```python
# Average change in temperature of every year
yearly_mean = df[year_cols].mean()
print(yearly_mean)
```

```python
# Average change in temperature of entire data 
global_avg_all_years = yearly_mean.mean()
print("Global Average change in temperature:",global_avg_all_years)
# Actual Global Change in temperature at present as compared to baseline temperature 
```

```python
# Every countries Average change in temperature over the period 
country_mean = df.set_index("Country Name")[year_cols].mean(axis=1)
print(country_mean)
```

```python
# Country with highest average rise in temperature over the year
highest_country = country_mean.idxmax()
highest_value = country_mean.max()
print(highest_country,highest_value)
```

```python
# Country with lowest average drop in temperature over the year
lowest_country = country_mean.idxmin()
lowest_value = country_mean.min()
print(lowest_country,lowest_value)
```

```python
## Number of countries with the temperature rise above 2 degree celcius 

year_cols = [col for col in df.columns if col.isdigit()]

# Rise relative to 1970 for each year
temp_change = df[year_cols].subtract(df["1970"], axis=0)

# The maximum rise for each country
df["max_rise"] = temp_change.max(axis=1)

# Countries with >2°C rise
countries_over_2 = df[df["max_rise"] > 2]["Country Name"]
print("Number of countries with >2°C rise:", len(countries_over_2))
print("\nCountries:\n", countries_over_2.tolist())
```

```python
future_years = list(range(2022, 2051))  
results = []
```

```python
### Prediction done using Linear regression ML model continuing with the same trend till the year 2050

for i, row in df.iterrows():
    country = row['Country Name']
    y = np.array([int(y) for y in year_cols]).reshape(-1, 1)
    values = row[year_cols].values

    # Training model
    X_train, X_test, y_train, y_test = train_test_split(y, values, test_size=0.2, random_state=42)
    lin_reg = LinearRegression()
    lin_reg.fit(X_train, y_train)

    # Predicting future
    future_X = np.array(future_years).reshape(-1, 1)
    future_preds = lin_reg.predict(future_X)

    # Storeing results 
    result = {"Country": country}
    for year, pred in zip(future_years, future_preds):
        result[str(year)] = pred

    results.append(result)
```

```python
pred_df = pd.DataFrame(results)
print(pred_df.head())
```

```python
pred_df
```

```python
# Average predicted temperature change form year 2022 to 2050 
mean_per_year = pred_df[[str(y) for y in future_years]].mean()
mean_df = mean_per_year.reset_index()
mean_df.columns = ["Year", "Mean_Temp_Change"]
print(mean_df)
```

```python
# Predicted highly affected country due to temperature change by the year 2050
max_row = pred_df.loc[pred_df["2050"].idxmax()]
highest_country = max_row["Country"]
highest_value = max_row["2050"]
print(highest_country,highest_value)
```

```python
# Predicted least affected country due to temperature change by the year 2050
min_row = pred_df.loc[pred_df["2050"].idxmin()]
lowest_country = min_row["Country"]
lowest_value = min_row["2050"]
print(lowest_country,lowest_value)
```

```python
## predicted number of countries with the temperature rise above 2 degree celcius by the year 2050 

baseline = df[["Country Name", "1970"]].rename(columns={"1970": "baseline"})
merged = pred_df.merge(baseline, left_on="Country", right_on="Country Name")
merged["rise_2050"] = merged["2050"] - merged["baseline"]
countries_over_2_2050 = merged[merged["rise_2050"] > 2]["Country"]

print("Number of countries with >2°C rise in 2050:", len(countries_over_2_2050))
print("\nCountries:\n", countries_over_2_2050.tolist())
```

```python
## Linegraph of average per year change in temperature of the given data followed by predicted data 

historical_years = [str(y) for y in range(1970, 2022)]
historical_mean = df[historical_years].mean()

future_mean = pred_df[[str(y) for y in future_years]].mean()

plt.figure(figsize=(18,6))
plt.plot(historical_mean.index, historical_mean.values, color="blue", label="Historical (1970–2021)")
plt.plot(future_mean.index, future_mean.values, color="red", label="Predicted (2022–2050)")

plt.axhline(2, color="gray", linestyle="--", label="2°C Threshold")
plt.title("Global Mean Temperature Change (1970–2050)")
plt.xlabel("Year")
plt.ylabel("Mean Temperature Change (°C)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```
